## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**
- Criar pelo menos 2 m√©todos de a√ß√£o para a classe principal.
- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 2

class Jogador:
    def __init__(self, nome, jogo, pontuacao):
        self.nome = nome
        self.jogo = jogo
        self.pontuacao = pontuacao

    def exibir_dados(self):
        print(f"Nome do Jogador: {self.nome}")
        print(f"Jogo: {self.jogo}")
        print(f"Pontua√ß√£o: {self.pontuacao}")
        print("-" * 20)

    def aumentar_pontuacao(self, pontos):
        self.pontuacao += pontos
        print(f"{self.nome} ganhou {pontos} pontos!")

    def zerar_pontuacao(self):
        self.pontuacao = 0
        print(f"A pontua√ß√£o de {self.nome} foi zerada.")

class JogadorPremium(Jogador):
    def __init__(self, nome, jogo, pontuacao, nivel):
        super().__init__(nome, jogo, pontuacao)
        self.nivel = nivel

    def bonus_vip(self):
        bonus = 500
        self.pontuacao += bonus
        print(f"{self.nome} recebeu {bonus} pontos de b√¥nus VIP por ser n√≠vel {self.nivel}!")

    def exibir_dados(self):
        super().exibir_dados()
        print(f"N√≠vel VIP: {self.nivel}")
        print("-" * 20)

jogador1 = Jogador("Alice", "Space Battle", 1500)
jogador2 = JogadorPremium("Bruno", "Space Battle", 2300, "Ouro")

jogador1.aumentar_pontuacao(200)
jogador1.zerar_pontuacao()
jogador2.bonus_vip()

jogador1.exibir_dados()
jogador2.exibir_dados()
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

```python
# Desenvolva aqui o c√≥digo da Parte 3

class Jogador:
    def __init__(self, nome, jogo, pontuacao):
        self.nome = nome
        self.jogo = jogo
        self.__pontuacao = pontuacao

    def get_pontuacao(self):
        return self.__pontuacao

    def set_pontuacao(self, nova_pontuacao):
        if nova_pontuacao < 0:
            print("A pontua√ß√£o n√£o pode ser negativa!")
        else:
            self.__pontuacao = nova_pontuacao

    def verificar_status(self):
        if self.__pontuacao >= 2000:
            return f"{self.nome} desbloqueou o status de Jogador Experiente!"
        else:
            return f"{self.nome} ainda n√£o alcan√ßou o status de Jogador Experiente."

    def exibir_dados(self):
        print(f"Nome do Jogador: {self.nome}")
        print(f"Jogo: {self.jogo}")
        print(f"Pontua√ß√£o: {self.get_pontuacao()}")
        print("-" * 20)

jogador1 = Jogador("Alice", "Space Battle", 1500)
jogador2 = Jogador("Bruno", "Space Battle", 2500)

jogador1.exibir_dados()
jogador2.exibir_dados()

jogador1.set_pontuacao(1800)
print(f"Nova pontua√ß√£o de {jogador1.nome}: {jogador1.get_pontuacao()}")

print(jogador1.verificar_status())
print(jogador2.verificar_status())
```

## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 4

import sqlite3

class Jogador:
    def __init__(self, nome, jogo, pontuacao):
        self.nome = nome
        self.jogo = jogo
        self.pontuacao = pontuacao

    def exibir_dados(self):
        return f"Nome do Jogador: {self.nome}, Jogo: {self.jogo}, Pontua√ß√£o: {self.pontuacao}"

class JogadorPremium(Jogador):
    def __init__(self, nome, jogo, pontuacao, nivel):
        super().__init__(nome, jogo, pontuacao)
        self.nivel = nivel

    def exibir_dados(self):
        return f"Nome do Jogador: {self.nome}, Jogo: {self.jogo}, Pontua√ß√£o: {self.pontuacao}, N√≠vel VIP: {self.nivel}"


def conectar_db():
    conn = sqlite3.connect("jogadores.db")
    return conn

def criar_tabela():
    conn = conectar_db()
    cursor = conn.cursor()
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS jogadores (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            nome TEXT,
            jogo TEXT,
            pontuacao INTEGER,
            nivel TEXT
        )
    """)
    conn.commit()
    conn.close()

def inserir_jogador(jogador):
    conn = conectar_db()
    cursor = conn.cursor()
    cursor.execute("""
        INSERT INTO jogadores (nome, jogo, pontuacao, nivel) 
        VALUES (?, ?, ?, ?)
    """, (jogador.nome, jogador.jogo, jogador.pontuacao, getattr(jogador, 'nivel', None)))
    conn.commit()
    conn.close()

def consultar_jogadores():
    conn = conectar_db()
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM jogadores")
    jogadores = cursor.fetchall()
    conn.close()
    return jogadores

def excluir_jogador(id):
    conn = conectar_db()
    cursor = conn.cursor()
    cursor.execute("DELETE FROM jogadores WHERE id = ?", (id,))
    conn.commit()
    conn.close()

criar_tabela()

jogador1 = Jogador("Alice", "Space Battle", 1500)
jogador2 = JogadorPremium("Bruno", "Space Battle", 2500, "Ouro")

inserir_jogador(jogador1)
inserir_jogador(jogador2)

print("Jogadores no banco de dados:")
for jogador in consultar_jogadores():
    print(jogador)

excluir_jogador(1)

print("\nJogadores ap√≥s exclus√£o:")
for jogador in consultar_jogadores():
    print(jogador)
```

```python
# Desenvolva aqui o c√≥digo da Parte 5

import sqlite3

class Jogador:
    def __init__(self, nome, jogo, pontuacao):
        """
        Inicializa os dados do jogador, incluindo nome, jogo e pontua√ß√£o.
        """
        self.nome = nome
        self.jogo = jogo
        self.pontuacao = pontuacao

    def exibir_dados(self):
        """
        M√©todo para exibir os dados do jogador. Pode ser sobrescrito nas subclasses.
        """
        return f"Nome do Jogador: {self.nome}, Jogo: {self.jogo}, Pontua√ß√£o: {self.pontuacao}"

    def aumentar_pontuacao(self, pontos):
        """
        M√©todo para aumentar a pontua√ß√£o do jogador.
        """
        self.pontuacao += pontos
        print(f"{self.nome} ganhou {pontos} pontos!")

    def zerar_pontuacao(self):
        """
        M√©todo para zerar a pontua√ß√£o do jogador.
        """
        self.pontuacao = 0
        print(f"A pontua√ß√£o de {self.nome} foi zerada.")

    def verificar_status(self):
        """
        M√©todo para verificar se o jogador atingiu a pontua√ß√£o m√≠nima para desbloquear um status especial.
        """
        if self.pontuacao >= 2000:
            return f"{self.nome} desbloqueou o status de Jogador Experiente!"
        else:
            return f"{self.nome} ainda n√£o alcan√ßou o status de Jogador Experiente."

class JogadorPremium(Jogador):
    def __init__(self, nome, jogo, pontuacao, nivel):
        """
        Inicializa o jogador premium, herdando da classe Jogador e adicionando o atributo 'nivel'.
        """
        super().__init__(nome, jogo, pontuacao)
        self.nivel = nivel

    def exibir_dados(self):
        """
        Sobrescreve o m√©todo exibir_dados da classe Jogador, incluindo o n√≠vel VIP do jogador premium.
        """
        return f"Nome do Jogador: {self.nome}, Jogo: {self.jogo}, Pontua√ß√£o: {self.pontuacao}, N√≠vel VIP: {self.nivel}"

    def bonus_vip(self):
        """
        M√©todo exclusivo da subclasse para conceder um b√¥nus de pontos para jogadores premium.
        """
        bonus = 500
        self.pontuacao += bonus
        print(f"{self.nome} recebeu {bonus} pontos de b√¥nus VIP por ser n√≠vel {self.nivel}!")

def conectar_db():
    """
    Conecta ao banco de dados SQLite e retorna a conex√£o.
    """
    return sqlite3.connect("jogadores.db")

def criar_tabela():
    """
    Cria a tabela 'jogadores' no banco de dados caso n√£o exista.
    """
    conn = conectar_db()
    cursor = conn.cursor()
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS jogadores (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            nome TEXT,
            jogo TEXT,
            pontuacao INTEGER,
            nivel TEXT
        )
    """)
    conn.commit()
    conn.close()

def inserir_jogador(jogador):
    """
    Insere um jogador no banco de dados.
    """
    conn = conectar_db()
    cursor = conn.cursor()
    cursor.execute("""
        INSERT INTO jogadores (nome, jogo, pontuacao, nivel) 
        VALUES (?, ?, ?, ?)
    """, (jogador.nome, jogador.jogo, jogador.pontuacao, getattr(jogador, 'nivel', None)))
    conn.commit()
    conn.close()

def consultar_jogadores():
    """
    Consulta todos os jogadores registrados no banco de dados.
    """
    conn = conectar_db()
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM jogadores")
    jogadores = cursor.fetchall()
    conn.close()
    return jogadores

def excluir_jogador(id):
    """
    Exclui um jogador do banco de dados pelo ID.
    """
    conn = conectar_db()
    cursor = conn.cursor()
    cursor.execute("DELETE FROM jogadores WHERE id = ?", (id,))
    conn.commit()
    conn.close()

criar_tabela()

jogador1 = Jogador("Alice", "Space Battle", 1500)
jogador2 = JogadorPremium("Bruno", "Space Battle", 2500, "Ouro")

inserir_jogador(jogador1)
inserir_jogador(jogador2)

print("Jogadores no banco de dados:")
for jogador in consultar_jogadores():
    print(jogador)

jogador1.aumentar_pontuacao(500)
jogador1.zerar_pontuacao()
print(jogador1.verificar_status())

jogador2.bonus_vip()
print(jogador2.verificar_status())

excluir_jogador(1)

print("\nJogadores ap√≥s exclus√£o:")
for jogador in consultar_jogadores():
    print(jogador)
```
